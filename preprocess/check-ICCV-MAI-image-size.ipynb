```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import shutil
from tqdm import tqdm
```

```python
covidx_dir = '/media/ubuntu/MyHDataStor3/datasets/COV19D/test'
print(covidx_dir)
```

```python
import glob
all_dirs = glob.glob(covidx_dir+"/*/*/")
print(all_dirs[:10])
```

```python
pbar = tqdm(total=len(all_dirs))
count = 0 

for d in all_dirs: 
    pbar.update()
    
    if os.path.isfile(d):
        continue 
    files = os.listdir(d) 
    for f in files:
        f_full = os.path.join(d,f)
        
        if "._" in f_full: 
            print(f)
            print(f_full)
            continue 
        
        img = cv2.imread(f_full, cv2.IMREAD_UNCHANGED)
        #print(img.shape)
        if img is None: 
            print("file {} has problem".format(f_full))
            input("dbg image file")  
            continue          
            
            
        if img.shape != (512,512):                
            print("file {} shape = {}".format(f_full,img.shape))                     
            img = cv2.resize(img,(512,512),cv2.INTER_LINEAR)
            cv2.imwrite(f_full,img)            
            count += 1   
            print("Resize count = {}".format(count))
```
