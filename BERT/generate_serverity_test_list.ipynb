```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import shutil
import tqdm 
```

```python
test_list_file = 'datasets/settings/covid2/test_feature_split1.txt'
lines = open(test_list_file,'r').readlines() 
lines = [x for x in lines if 'part_2_test_set_ECCV_22' in x]
print(lines[:5])
```

```python
test_list_file = 'datasets/settings/covid2/test_feature_split1.txt'
fp = open(test_list_file,'w')
for l in lines: 
    l_split = l.split(' ')
    l_split[1] = '1' 
    l_new = ' '.join(l_split) 
    print(l_new)        
    fp.write(l_new)
fp.close() 
```

```python
train_list_file = 'datasets/settings/covid2/train_feature_split1.txt'
train_lines = open(train_list_file,'r').readlines() 
print(train_lines[:5])

val_list_file = 'datasets/settings/covid2/val_feature_split1.txt'
val_lines = open(val_list_file,'r').readlines() 
print(val_lines[:5])

```

```python
split = 5 

train_list_file = 'datasets/settings/covid/train_feature_split{}.txt'.format(split)
fp = open(train_list_file,'w')

for l in train_lines: 
    l_split = l.split(' ')
    
    if int(l_split[1]) == split-1:        
        l_split[1] = 'covid'
    else: 
        l_split[1] = 'non-covid'
    l_new = ' '.join(l_split)     
    fp.write(l_new) 
    
fp.close()
    
val_list_file = 'datasets/settings/covid/val_feature_split{}.txt'.format(split)
fp = open(val_list_file,'w')

for l in val_lines: 
    l_split = l.split(' ')
    
    if int(l_split[1]) == split-1:        
        l_split[1] = 'covid'
    else: 
        l_split[1] = 'non-covid'
    l_new = ' '.join(l_split)     
    fp.write(l_new) 
    
fp.close()
        
        
        
        

    
    
    

```
